```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
data1 = pd.read_pickle('../Data/total_sales_09-10_cleaned.pkl')
data1.head(10)
```

```python
data2 = pd.read_pickle('../Data/total_sales_10-11_cleaned.pkl')
data2.head(10)
```

```python
data1_byday = data1.groupby('date', as_index = False).sum()
data1_byday = data1_byday.drop(columns = ['hour'])
data1_byday.head()
```

```python
plt.figure(figsize=(15,10))
plt.plot(data1_byday['date'],data1_byday['total_amount'])
plt.ylabel('Total sales')
plt.xlabel('Time t by day')
# plt.legend(['Icream Daily Production'])
plt.grid(True)

locator = mdates.MonthLocator()  # every month
fmt = mdates.DateFormatter('%b')
X = plt.gca().xaxis
X.set_major_locator(locator)
X.set_major_formatter(fmt)
```
